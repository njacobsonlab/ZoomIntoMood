```python
def loop_through_imgs(result_dir, final_result_dir, df_csv_file, final_df_csv_file, wip):    
    # Construct path to current image directory based on whether it's an interviewer or participant image directory
    curr_dir = in_dir + pid + "/"
    if int_bool: 
        curr_img_dir = curr_dir + "interviewer_images/"
    else: 
        curr_img_dir = curr_dir + "participant_images/"

    # Initialize DataFrame to None and set starting index based on whether the work-in-progress csv exists
    df = None
    if wip:
        start_ind = pd.read_csv(df_csv_file).index[-1] + 1
        print_t("EXISTS, START INDEX:", start_ind)
    else:
        start_ind = 0
        print_t("DOES NOT EXIST, START INDEX:", start_ind)        
        
    # Loop through images in the current image directory
    if int_bool:
        file_prefix = "int_" + pid + "_"
    else: 
        file_prefix = "part_" + pid + "_"
    file_suffix = ".jpg"
    img_files = [f for f in os.listdir(curr_img_dir) if f.startswith(file_prefix) and f.endswith(file_suffix) and int(f.split("_")[-1].split(".")[0]) >= start_ind]

    # Sort files by integer value of x
    sorted_files = sorted(img_files, key=lambda f: int(f.split("_")[-1].split(".")[0]))
    
    for img_name in sorted_files:
        img_file = curr_img_dir + img_name 
        i = int(img_file.split("_")[-1].split(".")[0])
        print_t("IMG", img_file.split("/")[-1].split(".")[0])

        # If image exits
        if os.path.exists(img_file):
            # Run OpenFace on the current image file and catch any exceptions that occur
            command = "singularity exec -B /dartfs-hpc/rc/lab/J/JacobsonLab/7._TrackingDepression/9._TheVisionProject/ /dartfs-hpc/rc/lab/J/JacobsonLab/7._TrackingDepression/9._TheVisionProject/open_face/of_lo.sif /dartfs-hpc/rc/lab/J/JacobsonLab/7._TrackingDepression/9._TheVisionProject/scripts/scripts3/of2.sh"
            command += " " + img_file + " " + final_result_dir
            try:
                subprocess.run(command, shell=True)
            except Exception as e:
                print_t("ERROR RUNNING OPENFACE", pid, ": ", e)
                
            # Append the processed results to the DataFrame and check if tracking has failed
            df, track = append_lists(final_result_dir, img_file, df)
            if track:
                break                
                
            # Append the current DataFrame to the CSV file every 500 images
            if (not i %500):
                df = append_csvs(i, df, df_csv_file)
#                 break
    
    # Append the final DataFrame to the CSV file and rename the file, marking it complete
    if df is not None:
        df = append_csvs(i, df, df_csv_file)
#     os.rename(df_csv_file, final_df_csv_file)    
```

```python
def create_dir(dir_name):
    try:
        if not os.path.exists(dir_name):
            os.makedirs(dir_name)
            print_t(dir_name, "CREATED")
            return True
        else: return False
    except OSError as e:
        print_t("ERROR CREATING DIR", dir_name, "\n", e)
        return None
```

```python
def print_t(*args):
    current_time = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    print(current_time + ":", *args)
    
def log_to_file(log_file_name, *args):
    current_time = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    with open(log_file_name, 'a') as log_file:
        log_file.write(current_time + ": " + " ".join(str(arg) for arg in args) + "\n")
```
